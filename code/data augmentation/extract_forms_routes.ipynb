```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('CIS_bdpm2.txt', sep='\t', header=None)
df.head()
```

```python
df.head()
```

```python
df.columns=["id", "drug", "form", "route", "autor", "proc", "statut", "date", "jsp1", "jsp2", "labo", "jsp3"]
```

```python
df = df[["drug", "form", "route"]]
```

```python
df.nunique()
```

```python
df = df.replace('È','é', regex=True)
df = df.replace(chr(8225), 'à', regex=True)
df = df.replace('Ë', 'è', regex=True)
df = df.replace('‚', 'â', regex=True)
df = df.replace('Ù', 'ô', regex=True)

formes = df["form"].unique()
routes = df["route"].unique()
```

```python
import re
formes = [re.sub(r" +"," ",x) for x in formes]
formes = [re.sub(r"^ ","",x) for x in formes]
formes = [re.sub(r" $","",x) for x in formes]
formes = [re.sub(r"ou$","",x) for x in formes]
formes = [re.sub(r"et$","",x) for x in formes]
formes
```

```python
routes = [re.sub(r" +"," ",x) for x in routes]
routes = [re.sub(r"^ ","",x) for x in routes]
routes = [route.split(";") for route in routes]
routes = [elem for sublist in routes for elem in sublist] #flatten
routes = {route: 0 for route in routes}
routes = list(routes.keys())
```

```python
sorted(routes)
```

```python
sorted(formes)
```

```python
with open("routes.txt", 'w') as output:
    for row in sorted(routes):
        output.write(str(row) + '\n')
```

```python
with open("forms.txt", 'w') as output:
    for row in sorted(formes):
        output.write(str(row) + '\n')
```
