```python
import pandas as pd
import numpy as np
import random as rd
```

```python
#dictionnaire associant à chaque label une liste générique pour la data augmentation
labels_aug = {
    "Drug": {"filename": "drugs"},
    "Form": {"filename": "forms"},
    "Route": {"filename": "routes"},
    "Dosage": {"words": ["12g"]}
}

for label in labels_aug:
    try:
        txt_file = open("data/" + labels_aug[label]["filename"] + ".txt", "r")
        file_content = txt_file.read()

        ls = file_content.split("\n")[:-1]
        txt_file.close()

        labels_aug[label]["words"] = ls
    except:
        continue
```

```python
#tokenizer
from transformers import CamembertTokenizer
tz = CamembertTokenizer.from_pretrained("camembert-base")

#tokenization des mots pour chaque label
for label in labels_aug:
    labels_aug[label]["tokenized"] = [tz.tokenize(word) for word in labels_aug[label]["words"]]
```

```python
#conversion des tokens en labels
for label in labels_aug:
    for (k, tokenized) in enumerate(labels_aug[label]["tokenized"]):
        labels_aug[label]["tokenized"][k] = [[tokenized[i], "B-" + label if i==0 else "I-" + label] for i in range(len(tokenized))]
```
